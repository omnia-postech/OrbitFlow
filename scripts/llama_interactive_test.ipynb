```python
from vllm import LLM, SamplingParams
import os
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
llm = LLM(
            model="/home/jongseop/.cache/huggingface/hub/models--meta-llama--Meta-Llama-3.1-8B-Instruct/snapshots/0e9e39f249a16976918f6564b8830bc894c89659",
            gpu_memory_utilization=0.38,
            max_model_len=500,
            tensor_parallel_size=1,
            distributed_executor_backend='ray',
            enforce_eager=True
        )
```

```python
prompts = ["A farmer has 17 sheep. All but 9 run away. Later, 5 more are found and returned. Then, 3 sheep are sold, and 2 are born. How many sheep does the farmer have now? Walk through the problem step-by-step, explaining your reasoning clearly before giving the final answer.", "Imagine Earth receives its first message from an alien civilization. The message is just one word: “Balance.” Write a short story from the perspective of a linguist trying to decode the meaning and intent behind the message. Keep the tone suspenseful and grounded in plausible science."]
samplings = [SamplingParams(temperature=0, max_tokens=10)]*2 # , logprobs=10)
output = llm.generate(prompts,samplings)
```
